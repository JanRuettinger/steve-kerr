```python
from youtubesearchpython import *
import pandas as pd
import numpy as np
import re
import os
```

```python
from youtubesearchpython import ChannelsSearch
channelsSearch = ChannelsSearch('Dr Eric Berg', limit = 10, region = 'US')
print(channelsSearch.result())
```

```python


# https://pypi.org/project/youtube-search-python/
channel_id = "UC3w193M5tYPJqF0Hi-7U-2g" # Get ID from ChannelsSearch
playlist = Playlist(playlist_from_channel_id(channel_id))
while playlist.hasMoreVideos:
    print('Getting more videos...')
    playlist.getNextVideos()
    print(f'Videos Retrieved: {len(playlist.videos)}')

```

```python
max_videos = 4840
video_links=pd.DataFrame(columns = ["title", "link", "id", "transcript"], index = range(max_videos))
num_v = 0
for v in playlist.videos:
    try:
        video_links.loc[num_v, "title"] = v["title"]
        video_links.loc[num_v, "link"] = v["link"]
        video_links.loc[num_v, "id"] = v["id"]
        num_v = num_v + 1
    except:
        print("Failed on %s", v['title'])
    if num_v > max_videos:
        break
```

```python
video_links
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
output_dir = '../../ericberg_index'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    
for v in range(video_links.shape[0]):
    
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_links.loc[v,"id"])
        video_links.loc[v,"transcript"] = [sen["text"] for sen in transcript]

         # Extract the title and content
        title = video_links.loc[v,"title"]
        content = video_links.loc[v,"transcript"]

        # Remove special characters from the title
        title = re.sub(r'[^\w\s]', '', title)

        # Replace spaces with underscores
        title = title.replace(' ', '_')

        # Create the filename
        filename = os.path.join(output_dir, f'{title}.txt')

        # Write the content to the file
        with open(filename, 'w') as file:
            file.write(" ".join(content))
    except:
        print("Failed to transcriber for ",video_links.loc[v,"title"] )

#     break
```
